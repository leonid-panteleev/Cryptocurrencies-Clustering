##### Построим кластера на DTW

```python
def kmeans_graph(data, max_clusters=10, seed=23, metric='euclidean'):
    # Главные метрики
    distortions = []
    silhouette = []
    range_cl = range(1, max_clusters+1)
    
    for k in tqdm(range_cl):
        kmeans_model = TimeSeriesKMeans(n_clusters=k, metric=metric, n_jobs=-1, max_iter=10, random_state=seed)
        kmeans_model.fit(data)
        distortions.append(kmeans_model.inertia_)
        if k > 1:
            silhouette.append(silhouette_score(data, kmeans_model.labels_))
    
    fig, ax1 = plt.subplots()

    ax2 = ax1.twinx()
    ax1.plot(range_cl, distortions, 'b-')
    ax2.plot(range_cl[1:], silhouette, 'r-')

    ax1.set_xlabel('# clusters')
    ax1.set_ylabel('Distortion', color='b')
    ax2.set_ylabel('Silhouette', color='r')

    plt.show()
```
